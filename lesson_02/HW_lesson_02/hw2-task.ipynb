#### Задание 3. Фунция предобработки

Напишите функцию, которая бы

* Удаляло идентификатор `AGREEMENT_RK`
* Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
* Что-то делала с пропусками
* Кодировала категориальные признаки

В результате, ваш датафрейм должен содержать только числа и не содержать пропусков!

```python
def clearTextAndMakeFloat(df_input, columnName):
    df_input[columnName] = df_input[columnName].map(lambda x: x.replace(',', '.')).astype('float')
    
```

```python
def preproc_data(df_input):
    df_output = df_input.copy()
    df_output = df_output.drop('AGREEMENT_RK', axis=1)
    df_output = df_output.drop('WORK_TIME', axis=1) # данные очень противоречивые начиная от того что 1300+ пустые и несколько показателей больше 100 лет!
    
    clearTextAndMakeFloat(df_output, 'PERSONAL_INCOME')
    clearTextAndMakeFloat(df_output, 'CREDIT')
    clearTextAndMakeFloat(df_output, 'FST_PAYMENT')
    clearTextAndMakeFloat(df_output, 'LOAN_AVG_DLQ_AMT')
    clearTextAndMakeFloat(df_output, 'LOAN_MAX_DLQ_AMT')
    df_output['PREVIOUS_CARD_NUM_UTILIZED'] = df_output.loc[df_output['PREVIOUS_CARD_NUM_UTILIZED'].isnull(), 'PREVIOUS_CARD_NUM_UTILIZED'] = 0
    
    # удаляем пропуски - все крутиться вокруг 1367+- записе с пустыми значеними в различных статусах
    df_output = df_output[~df_output['ORG_TP_FCAPITAL'].isnull()]
    df_output = df_output[~df_output['GEN_INDUSTRY'].isnull()]
    df_output = df_output[~df_output['GEN_TITLE'].isnull()]
    df_output = df_output[~df_output['JOB_DIR'].isnull()]
  
    # расширяем
    df_output = pd.get_dummies(df_output, columns=['EDUCATION','ORG_TP_FCAPITAL','MARITAL_STATUS','GEN_INDUSTRY','GEN_TITLE','ORG_TP_STATE','REG_ADDRESS_PROVINCE','FACT_ADDRESS_PROVINCE','POSTAL_ADDRESS_PROVINCE','TP_PROVINCE','REGION_NM','FAMILY_INCOME'])
    df_output = pd.get_dummies(df_output, columns=['JOB_DIR'])
    
    return df_output
```

```python
df_preproc = df.pipe(preproc_data)
df_preproc.head().T
```

#### Задание 5. Обучение и оценка качества разных моделей

```python
from sklearn.cross_validation import train_test_split
# test_size=0.3, random_state=42

## Your Code Here

x_train, x_testSplit, y_train, y_testSplit = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
print('X_train', x_train.shape)
print('X_testSplit', x_testSplit.shape)
print('Y_train', y_train.shape)
print('Y_testSplit', y_testSplit.shape)
```

```python
def runModel(_model, _x_train, _y_train, _x_testSplit):
    _model.fit(_x_train, _y_train)
    _predict = _model.predict(_x_testSplit)
    _predict_proba = model.predict_proba(_x_testSplit)
    return _predict,_predict_proba
```

```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
# Постепенно мы узнаем, что они делают а сейчас учимся понимать какой работает качественнее
#     fit, 
#     predict, 
#     predict_proba

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

## Your Code Here
#RandomForestClassifier
modelDTC = DecisionTreeClassifier()
modelRFС = RandomForestClassifier()
modelLR = LogisticRegression()
predictDTC, predictDTC_proba = runModel(modelDTC, x_train, y_train, x_testSplit)
predictRFC, predictFCC_proba = runModel(modelRFС, x_train, y_train, x_testSplit)
predictLR, predictLR_proba = runModel(modelLR, x_train, y_train, x_testSplit)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
def calcMetrics(_y_test, _predict):
    precision_score_local = precision_score(_y_test, _predict)
    recall_score_local = recall_score(_y_test, _predict)
    accuracy_score_local = accuracy_score(_y_test, _predict)
    f1_score_local = f1_score(_y_test, _predict)
    return precision_score_local, recall_score_local, accuracy_score_local, f1_score_local
```

```python
def printMetrics(modelName, _y_test, _predict, _predict_proba):
    print('Method name:', modelName)
    precision_score_local, recall_score_local, accuracy_score_local, f1_score_local = calcMetrics(_y_test, _predict)
    print('precision', precision_score_local)
    print('recall', recall_score_local)
    print('accuracy', accuracy_score_local)
    print('f1', f1_score_local)
    print('predict_proba', _predict_proba[:,1])
    print(classification_report(_y_test, _predict))
    print('='*10)
```

```python
# Посчитать метрики стандартные
# accuracy, precision, recall

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

## Your Code Here
printMetrics('DecisionTreeClassifier', y_testSplit, predictDTC, predictDTC_proba)
printMetrics('RandomForestClassifier', y_testSplit, predictRFC, predictFCC_proba)
printMetrics('LogisticRegression', y_testSplit, predictLR, predictLR_proba)
```

```python
precision_score_DTS, recall_score_DTS, accuracy_score_DTS, f1_score_DTS = calcMetrics(y_testSplit, predictDTC)
precision_score_RFC, recall_score_RFC, accuracy_score_RFC, f1_score_RFC = calcMetrics(y_testSplit, predictRFC)
precision_score_LR, recall_score_LR, accuracy_score_LR, f1_score_LR = calcMetrics(y_testSplit, predictLR)

score_DTS = [precision_score_DTS, recall_score_DTS, accuracy_score_DTS, f1_score_DTS]
score_RFC = [precision_score_RFC, recall_score_RFC, accuracy_score_RFC, f1_score_RFC]
score_LR = [precision_score_LR, recall_score_LR, accuracy_score_LR, f1_score_LR]
```

```python
# Визуалищировать эти метрики всех моделей на одном графике (чтоб визуально сравнить)
# Возможно вас удивит качество! Но задача подобрана специально ;) Такое качество тоже бывает

## Your Code Here
%matplotlib inline
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))
#ax.set_xticklabels('precision', 'recall', 'accuracy', 'f1')
pos = [10, 11, 12, 13]

width = 0.25 
plt.bar(pos, score_DTS, width)
plt.bar([p + width for p in pos], score_RFC, width)
plt.bar([p + width*2 for p in pos], score_LR, width)

plt.legend(['DecisionTreeClassifier', 'RandomForestClassifier', 'LogisticRegression'], loc='upper left')
ax.set_ylabel('Score')
ax.set_title('Compare method metrics')
ax.set_xticks([p + 1.5 * width for p in pos])
ax.set_xticklabels(['precision', 'recall', 'accuracy', 'f1'])

plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, 1] )
plt.grid()
plt.show()
```
